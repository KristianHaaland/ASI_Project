### Plotting the results of NN training

This notebook is intended for plotting the NN test error results after training using Kaggle

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
simpleNN_df = pd.read_csv("Results/simpleNN/test_error_simpleNN.csv")
simpleNN_momentum_df = pd.read_csv("Results/simpleNN_momentum/test_error_simpleNN_momentum.csv")
bayesianNN_df = pd.read_csv("Results/bayesianNN/test_error_bayesianNN.csv")

simpleNN = simpleNN_df.iloc[:, 1]
simpleNN_momentum = simpleNN_momentum_df.iloc[:, 1]
bayesianNN = bayesianNN_df.iloc[:, 1]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(simpleNN[5::10], linestyle='-.', color = 'r', label='SGD')
plt.plot(simpleNN_momentum[5::10], marker='v', markerfacecolor='none', label='SGD with momentum')
plt.plot(bayesianNN[5::10], marker='x', markerfacecolor='none', label='SGHMC')
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Test Error', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
